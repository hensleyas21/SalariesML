```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data/sleep_cycle_productivity.csv')
df.describe()
```

```python
for col in df.columns:
    plot = sns.histplot(x=df[col])
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```
